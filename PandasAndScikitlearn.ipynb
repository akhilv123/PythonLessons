## Handle missing data

```python
df1 = pd.DataFrame({"one": pd.Series([1,2,3],index=['a','b','c']),"two":pd.Series([1,2,3,4], index = ['a','b','c','d'])})
df1
df1["three"]=df1["one"]*df1["two"]
df1
df1["four"]=[2,3,4,5]
df1
```

```python
df1.loc['b',['two','four']] = np.nan
df1.reset_index(drop = True, inplace = True)
df1
```

```python
# detect NaN
df1.notna()
```

```python
# detect NaN
df1.isna()
```

```python
# delete rows or columns with NaN
df1.dropna(axis=0)
```

```python
df1.dropna(axis=1)
```

```python
# Insert Scalar values to missing locations
df1.fillna(0)
```

```python
# Filling missing locations with interpolation
df1.interpolate()
```

```python
# Various interpolation method available in scipy
df1.interpolate(method='spline',order=1)
```

```python
# Multiple levels and reshaping by stack
tuples = list(
     zip(
          *[
              ['bar','bar','baz','baz','foo','foo','qux','qux'],
              ['one','two','one','two','one','two','one','two']
          ]
     )
)
index = pd.MultiIndex.from_tuples(tuples, names=['first','second'])
df1 = pd.DataFrame(np.random.randn(8,2),index=index,columns=["A","B"])
df1
```

```python
df1.stack()
```

```python
df1.stack().unstack()
```

## Statistics

Pandas provide methods to calculate descriptive statistics on Series and DataFrame

```python
df = pd.DataFrame({"one": pd.Series([1,2,3], index = ["a","b","c"]),"two":pd.Series([4,5,6], index = ["a","b","c",])})
df
```

```python
# Mean value over one axis
df.mean(axis=0)
```

```python
df.mean(axis=1)
```

```python
x= iris.iloc[:,2:4].to_numpy()
y= iris.species.to_numpy()
x[:5],y[:5]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3,random_state=1, stratify=y)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
# StandardScaler takes the mean and standard deviation of the input and normalize the values into a Gaussian Distribution
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
f,ax = plt.subplots(2,1, figsize=(6,8))
ax[0].hist(x_train, bins=15)
ax[1].hist(x_train_std, bins=15)
plt.show()
```
